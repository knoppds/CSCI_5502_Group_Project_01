```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
intel_url = 'https://en.wikipedia.org/wiki/List_of_Intel_processors'
request = requests.get(intel_url)
intel_soup = BeautifulSoup(request.content, 'html.parser')

generation_14_tabel = intel_soup.find_all('table', class_='wikitable sortable')[0]
generation_14_CPUs = generation_14_tabel.find_all('a', class_='external text')
#strip the text from the html tags
generation_14_CPUs = [cpu.text for cpu in generation_14_CPUs]
generation_14_CPUs

generation_13_tabel = intel_soup.find_all('table', class_='wikitable sortable')[2]
generation_13_CPUs = generation_13_tabel.find_all('td')
#strip the text from the html tags
generation_13_CPUs = [cpu.text for cpu in generation_13_CPUs]
#keep only elements that start with 13
generation_13_CPUs = [cpu for cpu in generation_13_CPUs if cpu.startswith('13')]
#remove the \n from the strings
generation_13_CPUs = [cpu.strip() for cpu in generation_13_CPUs]

generation_12_table = intel_soup.find_all('table', class_='wikitable sortable')[3]
generation_12_CPUs = generation_12_table.find_all('td')
# Strip the text from the HTML tags
generation_12_CPUs = [cpu.text for cpu in generation_12_CPUs]
# Keep only elements that start with 12 or G
generation_12_CPUs = [cpu for cpu in generation_12_CPUs if cpu.startswith('12') or cpu.startswith('G')]
# Remove the '\n' from the strings
generation_12_CPUs = [cpu.strip() for cpu in generation_12_CPUs]
#remove elements with less then 4 characters
generation_12_CPUs = [cpu for cpu in generation_12_CPUs if len(cpu) > 4]
#remove elements with MB at the end
generation_12_CPUs = [cpu for cpu in generation_12_CPUs if not cpu.endswith('MB')]

generation_11_table = intel_soup.find_all('table', class_='wikitable sortable')[4]
generation_11_CPUs = generation_11_table.find_all('td')
# Strip the text from the HTML tags
generation_11_CPUs = [cpu.text for cpu in generation_11_CPUs]
# Keep only elements that start with 11
generation_11_CPUs = [cpu for cpu in generation_11_CPUs if cpu.startswith('11')]
# Remove the '\n' from the strings
generation_11_CPUs = [cpu.strip() for cpu in generation_11_CPUs]

generation_10_table = intel_soup.find_all('table', class_='wikitable center sortable mw-collapsible')[2]
generation_10_CPUs = generation_10_table.find_all('td')
# Strip the text from the HTML tags
generation_10_CPUs = [cpu.text for cpu in generation_10_CPUs]
# Keep only elements that start with 10
generation_10_CPUs = [cpu for cpu in generation_10_CPUs if cpu.startswith('10')]
# Remove the '\n' from the strings
generation_10_CPUs = [cpu.strip() for cpu in generation_10_CPUs]
#remove elements with less then 4 characters
generation_10_CPUs = [cpu for cpu in generation_10_CPUs if len(cpu) > 4]

generation_9_table = intel_soup.find_all('table', class_='wikitable center sortable mw-collapsible')[4]
generation_9_CPUs = generation_9_table.find_all('td')
# Strip the text from the HTML tags
generation_9_CPUs = [cpu.text for cpu in generation_9_CPUs]
# Keep only elements that start with 9
generation_9_CPUs = [cpu for cpu in generation_9_CPUs if cpu.startswith('9')]
# Remove the '\n' from the strings
generation_9_CPUs = [cpu.strip() for cpu in generation_9_CPUs]
#remove elements with less then 4 characters
generation_9_CPUs = [cpu for cpu in generation_9_CPUs if len(cpu) > 3]

generation_8_table = intel_soup.find_all('table', class_='wikitable center sortable mw-collapsible')[5]
generation_8_CPUs = generation_8_table.find_all('td')
# Strip the text from the HTML tags
generation_8_CPUs = [cpu.text for cpu in generation_8_CPUs]
# Keep only elements that start with 8
generation_8_CPUs = [cpu for cpu in generation_8_CPUs if cpu.startswith('8')]
# Remove the '\n' from the strings
generation_8_CPUs = [cpu.strip() for cpu in generation_8_CPUs]
#remove elements with less then 4 characters
generation_8_CPUs = [cpu for cpu in generation_8_CPUs if len(cpu) > 3]

generation_7_table = intel_soup.find_all('table', class_='wikitable center sortable mw-collapsible')[7]
generation_7_CPUs = generation_7_table.find_all('td')
# Strip the text from the HTML tags
generation_7_CPUs = [cpu.text for cpu in generation_7_CPUs]
# Keep only elements that start with 7
generation_7_CPUs = [cpu for cpu in generation_7_CPUs if cpu.startswith('7') or cpu.startswith('G')]
# Remove the '\n' from the strings
generation_7_CPUs = [cpu.strip() for cpu in generation_7_CPUs]
#remove elements with less then 4 characters
generation_7_CPUs = [cpu for cpu in generation_7_CPUs if len(cpu) > 3]


#combine all into a single dataframe with a column for the generation
generation_14 = pd.DataFrame(generation_14_CPUs, columns=['CPU'])
generation_14['Generation'] = 14
generation_14['Release'] = 2023
generation_13 = pd.DataFrame(generation_13_CPUs, columns=['CPU'])
generation_13['Generation'] = 13
generation_13['Release'] = 2022
generation_12 = pd.DataFrame(generation_12_CPUs, columns=['CPU'])
generation_12['Generation'] = 12
generation_12['Release'] = 2021
generation_11 = pd.DataFrame(generation_11_CPUs, columns=['CPU'])
generation_11['Generation'] = 11
generation_11['Release'] = 2020
generation_10 = pd.DataFrame(generation_10_CPUs, columns=['CPU'])
generation_10['Generation'] = 10
generation_10['Release'] = 2019
generation_9 = pd.DataFrame(generation_9_CPUs, columns=['CPU'])
generation_9['Generation'] = 9
generation_9['Release'] = 2018
generation_8 = pd.DataFrame(generation_8_CPUs, columns=['CPU'])
generation_8['Generation'] = 8
generation_8['Release'] = 2017
generation_7 = pd.DataFrame(generation_7_CPUs, columns=['CPU'])
generation_7['Generation'] = 7
generation_7['Release'] = 2016
#combine all dataframes into a single dataframe
intel_processors = pd.concat([generation_14, generation_13, generation_12, generation_11, generation_10, generation_9, generation_8, generation_7])
intel_processors

#save the dataframes to a csv file
intel_processors.to_csv('Intel_Processors.csv', index=False)
```

```python
#load in CPU data
CPU_data = pd.read_csv('CPU_data.csv')
#split the CPU column into two columns, one for the brand and one for the model. The brand is the first word in the CPU column
CPU_data[['Brand', 'Model']] = CPU_data['CPU Name'].str.split(n=1, expand=True)
#dropping the CPU Name column and move the Brand and Model columns to the front
CPU_data = CPU_data.drop(columns=['CPU Name'])
CPU_data = CPU_data[['Brand', 'Model', 'Core Count', 'Performance Core Clock', 'Boost Clock', 'TDP', 'Integrated Graphics', 'Sockets Supported', 'Rating', 'Rating Count', 'Price']]
# put a new column in the dataframe for the generation of the CPU and fill it with NaN
CPU_data['Release Year'] = None
```

```python
#do the above function but it needs to have either 'Core', 'Pentium', or 'Celeron' in the model name
def find_release_year(Model):
    if 'Core' in Model or 'Pentium' in Model or 'Celeron' in Model:
        for index, row in intel_processors.iterrows():
            cpu_partial = row['CPU'][:4]
            if cpu_partial in Model:
                return row['Release']
        
    elif 'Ryzen' in Model or 'PRO' in Model:
        for index, row in amd_processors.iterrows():
            cpu_partial = row['CPU'][:4]
            if cpu_partial in Model:
                return row['Release']    
    else:
        return None    

# Apply the function to each CPU Name in CPU_df
CPU_data['Release Year'] = CPU_data['Model'].apply(find_release_year)
# Remove any rows with NaN Release Year
CPU_data = CPU_data.dropna(subset=['Release Year'])

# Remove any rows with 2 or less Core Count
CPU_data = CPU_data[CPU_data['Core Count'] > 2] 
#Remove any rows with NaN in the Price column
CPU_data = CPU_data.dropna(subset=['Price'])

#save the dataframes to a csv file
CPU_data.to_csv('CPU_Data_Release_Year.csv', index=False)
```
